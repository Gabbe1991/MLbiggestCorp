```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("company_data.csv")
for key in df.keys():
    print(key)
```

```python
df_copy = df.copy()

# Sort the DataFrame by 'Market Cap' and 'Revenue'
df_copy_market_cap = df_copy.sort_values(by='Market Cap', ascending=False).head(10)
df_copy_revenue = df_copy.sort_values(by='Revenue', ascending=False).head(10)

# Plot the data for top earners by Market Cap
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(df_copy_market_cap['Company Name'], df_copy_market_cap['Market Cap'], color='blue')
plt.xlabel('Company')
plt.ylabel('Market Cap')
plt.title('Top Companies by Market Cap')
plt.xticks(rotation=45)

# Plot the data for top earners by Revenue
plt.subplot(1, 2, 2)
plt.bar(df_copy_revenue['Company Name'], df_copy_revenue['Revenue'], color='green')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Top Companies by Revenue')
plt.xticks(rotation=45)

plt.tight_layout(pad=3.0)
plt.show()
```
